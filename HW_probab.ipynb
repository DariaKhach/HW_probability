Монетку подбрасывают 15 раз, найдите вероятность выпадения простого числа орлов.

```python
import math as m
import random
import numpy as np
```

```python
kl = []
p=0.5
n=15
l = list(range(1, n + 1))

def simple_prime(x):
    if x == 1:
        return False
    ki = list(range(2, int(m.sqrt(x)) + 1))
    for j in ki:
        if x % j == 0:
            return False
    return True
    
```

```python
for i in l:
    if simple_prime(i):
        kl.append(i)
```

```python
all_vae =[]
for h in kl:
    C_kn = m.factorial(n)/(m.factorial(h) * m.factorial(n - h))
    probability_of_success = C_kn * (p**h) * (1 - p)**(n - h)  
    all_vae.append(probability_of_success)
```

```python
sum(all_vae)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from scipy import stats
xk = np.arange(10)
pk = (0.1, 0.17, 0.004, 0.03, 0.106, 0.1, 0.12, 0.1, 0.1, 0.17)
custm = stats.rv_discrete(name='custm', values=(xk, pk))
```

```python
R = custm.rvs(size=10000)
R
```

```python
h = random.sample(list(R), 1000)
h
```

```python
plt.hist(h, bins = 100)
```

```python
print(np.mean(h))
print(np.var(h))
print(np.median(h))
```
